VISUALISASI DATASET

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('results.csv')
df.head(3)
```

```python
df.keys()
```

```python
df.neutral.value_counts()
```

GRAFIK PIE
Grafik Pie sangat cocok digunakan pada data categorical dan juga grafik ini bisa digunakan untuk melihat distribusi.

```python
plt.figure(figsize=(8,4))
plt.title('GRAFIK PIE')
plt.pie(df['neutral'].value_counts(), labels=['False','True', ],autopct='%.1f%%', explode=(0.1,0))
plt.legend(loc='lower left')
plt.show()
```

GRAFIK HEATMAP 
merupakan grafik visualisasi atau pemetaan dengan menampilkan data dengan representasi warna yang berbeda-beda. Pada grafik Heatmap semakin tinggi angka suatu kelompok data maka warnanya akan semakin gelap dan disimbolkan dengan warna merah, grafik ini biasa digunakan untuk menunjukkan korelasi antara variabel.

```python
sns.heatmap(data=df.corr(), cmap='pink')
plt.title('GRAFIK HEATMAP')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.show()
```

```python
df.info()
```

```python
df['home_score'].value_counts().plot.bar()
plt.title('GRAFIK BATANG')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.xticks(rotation=50)
plt.show()
```

GRAFIK BOXPLOT
Grafik Boxplot merupakan ringkasan distribusi sampel yang disajikan secara grafis yang bisa menggambarkan bentuk distribusi data (skewness), ukuran tendensi sentral dan ukuran penyebaran (keragaman) data pengamatan

```python
df.corr()
```

```python
plt.boxplot(df[['home_score','away_score','neutral']], labels=['Skor Awal','Skor Akhir','Netral'])
plt.title('GRAFIK BOXPLOT')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.show()
```
